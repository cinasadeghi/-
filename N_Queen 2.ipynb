# Importing

```python
from random import randint as rnd
from random import shuffle
import matplotlib.pyplot as plt
```

# Set parameter

```python
N = 64
Population_size = 200
MR = 0.8 #mutation rate
Epock=200 #number of repeat for algorithm (number of generations)
```

# Initial population Function

```python
def init_population(n,ps):
    population_list=[]
    for i in range(ps):
        member=[]
        for j in range(n):
            member.append(rnd(0,n-1))
        population_list.append(member+[None])
    return population_list
```

# Cross-Over Function

```python
def cross_over(population_list,n,ps):
    for i in range(0,ps,2):
        child1=population_list[i][:n//2]+population_list[i+1][n//2:n]+[None]
        child2=population_list[i+1][:n//2]+population_list[i][n//2:n]+[None]
        population_list.append(child1)
        population_list.append(child2)
    return population_list
```

# Mutation 

```python
def mutation(population_list,n,ps,mr):
    choosen_one = list(range(Population_size,Population_size*2))
    shuffle(choosen_one)
    choosen_one=choosen_one[:int(ps*mr)]
    
    for i in choosen_one:
        cell=rnd(0,n-1)
        val=rnd(0,n-1)
        population_list[i][cell]=val
    return population_list
```

# Fitness function

```python
def fitness(population_list,n):
    length=len(population_list)
    for i in range(length):
        conflict=0
        for j in range(n):
            for k in range(j+1,n):
                #column
                if population_list[i][j]==population_list[i][k]:
                    conflict+=1
                #diagonal
                if abs(j-k)==abs(population_list[i][j]-population_list[i][k]):
                    conflict+=1
                
        population_list[i][-1]=conflict
    return population_list
```

# Presentation Function

```python
def showResult(sol, n):
    fig, ax = plt.subplots()
    for i in range(n + 1):
        plt.plot([0, 2 * n], [i * 2, i * 2], color='black')
        plt.plot([i * 2, i * 2], [0, 2 * n], color='black')
    for i in range(n):
        plt.scatter([i * 2 + 1], [sol[i] * 2 + 1], color='red', marker='o')
    ax.set_aspect('equal', adjustable='box')

    plt.show()
```

# Main

```python
current_population = init_population(N,Population_size)

current_population = fitness(current_population,N)
current_population = sorted(current_population, key=lambda x:x[-1])
if current_population[0][-1]==0:
        print("Solution is found in the initial population stage: "
              ,current_population[0])
        showResult(current_population[0],N)
else:      
    for i in range(Epock):
        current_population = cross_over(current_population,N,Population_size)
        current_population = mutation(current_population,N,Population_size,MR)
        current_population = fitness(current_population,N)
        current_population = sorted(current_population, key=lambda x:x[-1])
        current_population = current_population[:Population_size]

        if current_population[0][-1]==0:
            print(i+1,"Solution Found: ",current_population[0])
            showResult(current_population[0],N)
            break
        else:
            print(i+1,"Best solution so far: ",current_population[0])
            showResult(current_population[0],N)
    else:
        print("Sorry,we couldnt find a solution!")
```
